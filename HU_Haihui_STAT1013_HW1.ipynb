## **Q3: Data visualization in seaborn**

- Q3.1: print overall histplot and violinplot of `bp_before` via `sns.histplot` and `sns.violinplot`

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = [10, 5]

sns.set()
```

```python
## Your solution to Q3.1
sns.histplot(data=df, x='bp_before', bins='auto')
plt.show()

sns.violinplot(data=df, x='bp_before')
plt.show()
```

```python
## Your solution to Q3.2
sns.violinplot(data=df, x='bp_before', y='agegrp')
plt.show()
```

```python
## Your solution to Q3.3
sns.violinplot(data=df, x='bp_after', y='agegrp', hue='sex')
plt.show()
```

```python
## Your solution to Q3.4

## Compare the distribution of "bp_before" and "bp_after".
sns.kdeplot(x=df['bp_before'])
sns.kdeplot(x=df['bp_after'])
plt.show
```

The blue cruve and orange cruve show the distribution of "bp_before" and "bp_after" respectively. The overall "bp_after" is lower than "bp_before".

```python
## Compare the bp_before based on different agegrp under the effect of gender
sns.violinplot(data=df, x='bp_before', y='agegrp', hue='sex')
plt.show()
```

## **Q4: Grouping Statistics via `groupby`**

> Please answer the question by using `groupby`

- Q4.1: Which `agegrp` results higher `bp_after` on average?

- Q4.2: Print the mean, std, min and max values for `bp_before` and `bp_after` based on different `agegrp`.

```python
## Your solution to Q4.1
print(df.groupby('agegrp')['bp_after'].mean())
```

From the above table, we know that the "60+" agegrp has a highest mean of bp_after, followed by "46-59" agegrp. "30-45" agegrp results a lowest bp_after on average.

```python
## Your solution to Q4.2
print(df.groupby('agegrp')[['bp_before','bp_after']].agg(['mean', 'std', 'min', 'max']))
```
